```python
from f1pitpred.loading import load_from_csv

figures_path = 'figures/'
years = range(2019, 2024)

data = load_from_csv(years)
data.reset_index(inplace=True)
```

```python
from f1pitpred.preprocessing import preprocess_pre_split

laps_df = preprocess_pre_split(data, target = 'tire', drop=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
laps_df.columns
```

```python
fig, axes = plt.subplots(figsize=(10, 6), ncols=2, sharey=True)

sns.kdeplot(data, x='LapTime', hue='PitStatus',
            common_norm=False,
            fill=True, palette='bright', legend=True, ax=axes[0])


sns.kdeplot(laps_df, x='LapTime', hue='is_pitting', 
            common_norm=False,
            fill=True, palette='bright', legend=True, ax=axes[1])

fig.suptitle('Lap time distribution by pit status and is_pitting')
fig.tight_layout()
fig.savefig(figures_path + "lap_time_distribution.svg", format="svg")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.displot(laps_df, x='Stint', hue='is_pitting',
            common_norm=False, stat='density',
            palette='bright', legend=True, discrete=True)
plt.title(f'Distribution of stint')
plt.ylabel('Density')
plt.show()
```

```python
def plot_distrib(col):
    plt.figure(figsize=(10, 6))
    sns.displot(laps_df, x=col, hue='is_pitting',
                common_norm=False, kind='kde',
                fill=True, palette='bright', legend=True)
    plt.title(f'Distribution of {col}')
    plt.ylabel('Density')
    plt.show()
```

```python
for col in laps_df.columns:
    # check column is not categorical
    if col not in ['LapTime', 'is_pitting'] and laps_df[col].dtype != 'object' and laps_df[col].dtype != 'bool' and laps_df[col].dtype != 'category':
        plot_distrib(col)
```
