```python
# imports
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# pandas display options
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

random_state = 42
np.random.seed(random_state)

base_path = 'data'
file_name = 'spanish_grand_prix.csv'
years = [2023]
```

```python
from f1pitpred.loading import load_from_csv

data = load_from_csv(years, base_path, file_name)
```

```python
race = data.query('DriverNumber == "44"')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10), sharex=True)


sns.lineplot(
    data=race,
    x='LapNumber',
    y='LapTime',
    ax=axes[0, 0]
)

sns.lineplot(
    data=race,
    x='LapNumber',
    y='TyreLife',
    ax=axes[0, 1]
)

sns.lineplot(
    data=race,
    x='LapNumber',
    y='NumberOfPitStops',
    ax=axes[1, 0]
)

axes[1, 0].set_yticks(range(0, race['NumberOfPitStops'].max() + 1))

sns.lineplot(
    data=race,
    x='LapNumber',
    y='Position',
    ax=axes[1, 1]
)

axes[1, 1].set_yticks(range(0, race['Position'].max() + 1))

axes[0, 0].set_title('Lap time')
axes[0, 1].set_title('Tyre life')
axes[1, 0].set_title('Number of pit stops')
axes[1, 1].set_title('Position')
plt.suptitle('Evolution of race data for Lewis Hamilton in the 2023 Spanish Grand Prix', fontsize=16)
plt.savefig('figures/hamilton_spanish_grand_prix.svg', format='svg')
```
